```python
import pandas as pd
import numpy as np
```

### Cleaning and combining the excle values.

```python
file_names = ['HLPCA-10209-2011_H14_census', 'HLPCA-10235-2011_H14_census', 'HLPCA-10240-2011_H14_census']
np1 = []
for file in file_names:
    # we need specific columns from the excel for displaying the data.
    df = pd.read_excel(file + '.xlsx', sheet_name='Sheet1', usecols="B, D, F, I, J, AW:AZ, BA:BJ, CG:CL, DX:DZ, EA:EI")
    # cleaning the data.
    df.drop_duplicates()
    df.dropna()
    if len(np1) > 0:
        np1 = np.concatenate((np1, df.to_numpy()[2:]))
    else:
        np1 = df.to_numpy()[2:]
```

### Renaming the columns for better use since there are merged columns.

```python
columns = ['State Name', 'District Name', 'Tehsil Name', 'Area Name', 'Rural/Urban', 'Number of Dwelling Rooms No exclusive room', 'Number of Dwelling Rooms One room', 'Number of Dwelling Rooms Two rooms', 'Number of Dwelling Rooms Three rooms', 'Number of Dwelling Rooms Four rooms', 'Number of Dwelling Rooms Five rooms', 'Number of Dwelling Rooms Six rooms and above', 'Household size 1', 'Household size 2', 'Household size 3', 'Household size 4', 'Household size 5', 'Household size 6-8', 'Household size 9+', 'Main Source of lighting Electricity', 'Main Source of lighting Kerosene', 'Main Source of lighting Solar energy', 'Main Source of lighting Other oil', 'Main Source of lighting Any other', 'Main Source of lighting No lighting', 'Asset Radio', 'Asset Televison', 'Asset Computer with internet', 'Asset Computer without internet', 'Asset Telephone landline only', 'Asset Telephone mobile only', 'Asset Telephone both', 'Asset Bicycle', 'Asset Scooter/ Motorcycle/Moped', 'Asset Car/ Jeep/Van', 'Asset Households with TV, Computer/Laptop, Telephone/mobile phone and Scooter/ Car', 'Asset None of the assets specified in col. 10 to 19',]
df = pd.DataFrame(np1, columns = columns)
df.head()
```
